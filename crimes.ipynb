Here based on the combined average mean value of the number of crimes for each column a district is classified as High Crime Area or Low crime area

```python
year_list=set(new_df['year'])
#state_list=list(new_df['state'])
#state_list=[x.lower() for x in state_list]
#state_list=set(state_list)
#print(state_list)
#print(state_list)
percent_crime_label=[]
per_crime=[]
state_list=[]
cols=['murder','aow','ka','dacoity','robbery','arson','hurt','poa_act','pcr_act','sc_crimes']
print(year_list)
for i in year_list:
    yearwise_df=new_df[new_df['year']==i]
    original_state_list=list(yearwise_df['state'])
    for state in original_state_list:
        if state not in state_list:
            state_list.append(state)
    #state_list=set(state_list)
    print(state_list)
    for j in state_list:
        statewise_df=yearwise_df[yearwise_df['state']==j]
        #print(statewise_df)
        total_df=statewise_df[statewise_df['district']=='TOTAL']
        for total_ind in total_df.index:
            total_row=total_df.loc[total_ind]
        #print(total_row)
        for ind in statewise_df.index:
            each_row=statewise_df.loc[ind]
            if each_row['district']!='TOTAL':
                #print('Each row',each_row)
                su=0
                row_sum=0
                for n in cols:
                    row_sum=row_sum+each_row[n]
                if row_sum>=10:
                    for m in cols:
                        #print('total_df',total_df[m])
                        #print('Each Row',each_row[m])
                        if(total_row[m]>0):
                            div=each_row[m]/total_row[m]
                            div=round(div*100)
                        elif(total_row[m]==0):
                            div=0
                        #print(div)
                        su=su+div
                    #print('sum',su)
                    mean_percent=int(su//10)
                    #print('Mean percent',mean_percent)
                    per_crime.append(mean_percent)
                    if(mean_percent>=0 and mean_percent<=10):
                        percent_crime_label.append('LOW CRIME')
                    elif(mean_percent>10 and mean_percent<100):
                        percent_crime_label.append('HIGH CRIME')
                    su=0
                    #print('=='*50)

                else:
                    mean_percent=row_sum
                    #print('Mean percent',mean_percent)
                    per_crime.append(mean_percent)
                    if(mean_percent>=0 and mean_percent<=10):
                        percent_crime_label.append('LOW CRIME')
                    elif(mean_percent>10 and mean_percent<100):
                        percent_crime_label.append('HIGH CRIME')
                    row_sum=0
                    #print('=='*50)
            
    print('*'*100)
    state_list=[]
#print(per_crime)
#print(len(yearwise_df))
#yearwise_df
    
```

# Normalization

```python
from sklearn import preprocessing
normalize_df=pd.DataFrame([])
non_normalize_df=pd.DataFrame([])
normalize_df=train_df.drop(columns=['state','district','year'])
non_normalize_df=train_df.drop(columns=['murder','aow','ka','dacoity','robbery','arson','hurt','poa_act','pcr_act','sc_crimes'])
#print(normalize_df)
x = normalize_df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
normalize_df = pd.DataFrame(x_scaled)
normalize_df.columns=['murder','aow','ka','dacoity','robbery','arson','hurt','poa_act','pcr_act','sc_crimes']
normalize_df
```

# Concatenating columns

```python
state=list(train_df['state'])
district=list(train_df['district'])
normalize_df.insert(0,'state',state,True)
normalize_df.insert(1,'district',district,True)
normalize_df
```

# One-hot Encoding of state and district column

```python
normalize_df = pd.concat([normalize_df,pd.get_dummies(normalize_df['state'], prefix='state')],axis=1)
normalize_df
# now drop the original 'country' column (you don't need it anymore)
#normalize_df.drop(['country'],axis=1, inplace=True)
```

```python
normalize_df = pd.concat([normalize_df,pd.get_dummies(normalize_df['district'], prefix='district')],axis=1)
normalize_df
```

# Training Dataset

```python
# now drop the original 'country' column (you don't need it anymore)
normalize_df.drop(['state'],axis=1, inplace=True)
normalize_df.drop(['district'],axis=1, inplace=True)
normalize_df
```

# target column in the form 0 - Low crime and 1 - High Crime

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
import warnings
warnings.filterwarnings('ignore')
```

# Splitting dataset, Training -70 and Testing -30

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
x_train, x_test, y_train, y_test = train_test_split(normalize_df,target_df, test_size=0.3, random_state=42)
```

# CART

```python
from sklearn.tree import DecisionTreeClassifier
model1=DecisionTreeClassifier()
model1.fit(x_train,y_train)
y_predict = model1.predict(x_test)
decisiontree_accuracy=accuracy_score(y_test, y_predict)
print(" Decision Tree Accuracy = ",decisiontree_accuracy)
predictions = cross_val_predict(model1, x_train, y_train, cv=3)
cfm=confusion_matrix(y_train, predictions)
print("Decision Tree Confusion Matrix",cfm)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import Perceptron
from sklearn.naive_bayes import GaussianNB
```

# Random Forest

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(x_train, y_train)
y_predict= random_forest.predict(x_test)
randomforest_accuracy=accuracy_score(y_test, y_predict)
print("Random Forest Accuarcy = ",randomforest_accuracy)
predictions = cross_val_predict(random_forest, x_train, y_train, cv=3)
cfm=confusion_matrix(y_train, predictions)
print("Random Forest Confusion Matrix",cfm)

```

# Logistic Regression

```python
logreg=LogisticRegression()
logreg.fit(x_train, y_train)
y_predict=logreg.predict(x_test)
logreg_accuracy=accuracy_score(y_test, y_predict)
print("Logistic Regression Accuarcy = ",logreg_accuracy)
predictions = cross_val_predict(logreg, x_train, y_train, cv=3)
cfm=confusion_matrix(y_train, predictions)
print("Logistic Regression Confusion Matrix",cfm)
```

# Perceptron

```python
perceptron=Perceptron(max_iter=5)
perceptron.fit(x_train, y_train)
y_predict=perceptron.predict(x_test)
perceptron_accuracy=accuracy_score(y_test, y_predict)
print("Perceptron Accuarcy = ",perceptron_accuracy)
predictions = cross_val_predict(perceptron, x_train, y_train, cv=3)
cfm=confusion_matrix(y_train, predictions)
print("Perceptron Confusion Matrix",cfm)
```

# Naive Bayes

```python
gaussian=GaussianNB() 
gaussian.fit(x_train, y_train)  
y_predict=gaussian.predict(x_test)
guassian_accuracy=accuracy_score(y_test, y_predict)
print("Guassian Accuarcy = ",guassian_accuracy)
predictions = cross_val_predict(gaussian, x_train, y_train, cv=3)
cfm=confusion_matrix(y_train, predictions)
print("Guassian Confusion Matrix",cfm)
```
